## Part I - Mathematical and Python Function

```python
#imports numpy and pandas Libraries
import numpy as np
import pandas as pd
```

```python
#Question 1

#reads the file housing_data.csv
housing_df =pd.read_csv('housing_data.csv')

#counts the number of observations
n = len(housing_df)

#changes the column names to price and predicted to user easier
housing_df.columns = ['sqft','price','predicted']

#variables to simplify code from housing_df
price = housing_df.price
predicted = housing_df.predicted

#function to calculate the RMSE
def rmse (price,predicted):
    #finds the difference between the predicted price & sale price
    difference = np.subtract(predicted, price)                 

    #sum of squared differences
    ss = np.sum(np.square(difference))

    #calculates RMSE
    result = np.sqrt(ss/n)

    #prints the results
    print(f"The root mean square error (RMSE) is: {result}")

#calls the rmse function
rmse(price,predicted)
```

```python
#Question 2

#function to calculate the MSE
def mse (price, predicted):
    #finds the difference between the predicted price & sale price
    difference = np.subtract(predicted, price)                 

    #sum of squared differences
    ss = np.sum(np.square(difference))
    
    #calculates MSE
    result = (ss/n)
    print (f"The mean absolute error (MAE) is : {result}")

#calls the mse function
mse(price,predicted)
```

```python
#Question 3

#reads the file mushroom_data.csv
mushroom_df = pd.read_csv('mushroom_data.csv')

#variables to simply code from mushroom_data.csv
actual = mushroom_df.actual
predicted = mushroom_df.predicted

#function to calculate the accurate predictions made
def accurate (actual,predicted):
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            correct += 1
    accuracy = correct / float(len(predicted)) * 100
    #prints the results
    print ("The predicted accuracy of the mushrooms is {0:.0f}%.".format(accuracy))

#calls the funtion to calculate accurate predictions
accurate(actual,predicted)          
```

```python
#Question 5

#reads the file car_data.csv
car_df =pd.read_csv('car_data.csv')

#plot
plt.scatter(car_df.weight, car_df.hwy_mpg)

#label axes and adds a title
plt.xlabel('Weight(lbs)')
plt.ylabel('Highway MPG')
plt.title('MPG vs. Weight')

# add the line 
slope = -0.01
intercept = 55
prediction_line = slope * car_df.weight + intercept
plt.plot(car_df.weight, prediction_line, 'r', label='Predicted Best fit line') 
plt.legend() 

#show plot
plt.show()
```

```python
#Question 6

#imports scikit-learn package
from sklearn import linear_model

#reads the file car_data.csv
car_df =pd.read_csv('car_data.csv')

#plot
plt.scatter(car_df.weight, car_df.hwy_mpg)

#label axes and adds a title
plt.xlabel('Weight(lbs)')
plt.ylabel('Highway MPG')
plt.title('MPG vs. Weight')

#find slope & y-intercept
reg = linear_model.LinearRegression()
reg.fit(car_df[['weight']],car_df.hwy_mpg)
m = reg.coef_
b = reg.intercept_

# add the line 
slope = m
intercept = b
actual_line = m * car_df.weight + b
plt.plot(car_df.weight, actual_line, 'r', label='Actual Best fit line') 
plt.legend() 

#show plot
plt.show()
```

```python
#Question 7

#finds the difference of the prediction line & actual line.
difference = np.subtract(prediction_line, actual_line)                 

#sum of squared differences
ss = np.sum(np.square(difference))

#calculates RMSE
result = np.sqrt(ss/2)

#prints the results
print(f"The root mean square error (RMSE) is: {result}")
print("The predicted line and actual line are do not have a good accuracy,even though they look similar.")
```
