### Rieman Sum Function

```python
#imports libraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Function to calculate right Riemann sum for given function
def right_sum (f, a, b, n):
    dx = (b - a) / n
    sum = 0
    x = a    

    for i in range(n):
        x_right = x + dx
        sum += f(x_right) * dx
        x = x_right        
    return sum
    
#Defines the mathematical function
def f(x):
    return x ** 2 + 3
     
    
#Identifies the two interval endpoints a and b, and the number of rectangles n
a = 0
b = 3
n = 4

#Prints the results of the Riemann sum
Rn = right_sum(f,a,b,n)
print("The right Riemann sum (Rn) is",Rn)  
```

```python
# Define the function
def f(x):
    return np.sqrt(4 - x**2)
    
# Define the interval and the number of rectangles
a = 0
b = 2
n = 4

# # Calculate the width of each rectangle
width = (b - a) / n

# Create the x-values for the rectangles
x = np.linspace(a, b, n+1)

# Calculate the heights of the rectangles
heights = f(x[1:])

# Plot the graph of the function
x_vals = np.linspace(a, b, 100)
plt.plot(x_vals, f(x_vals), 'b-', label='f(x) = sqrt(4 - x^2)')

# Plot the rectangles
for i in range(n):
    plt.bar(x[i], heights[i], width, alpha=0.3, align='edge', edgecolor='black')

# Set the x-axis and y-axis labels
plt.xlabel('x')
plt.ylabel('f(x)')

# Set the plot title
plt.title('Graph of f(x) and Riemann Sum')

# Set the plot limits
plt.xlim(a, b)
plt.ylim(0, 2)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Define the function
def f(x):
    return np.sqrt(4 - x ** 2)

# Define the interval [a, b]
a = 0
b = 2
n = 1000

# Function to calculate the area under the curve
def a_u_c (f, a, b, n):
    h = (b - a) / n
    sum_f = f(a) + f(b)
    for i in range(1, n):
        sum_f += 2 * f(a + i * h)
    integral = (h / 2) * sum_f
    return integral

#Prints the results
area = a_u_c(f, a, b, n)
print("Approximate area under the curve:", area)
```

The area under the curve is called definite integrals or exact area.
