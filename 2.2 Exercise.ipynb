## US Population Growth

```python
#imports libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


#reads the file us_pop_data.csv
pop_df =pd.read_csv('us_pop_data.csv')
```

### a)

```python
#add column that tells the total of years since 1790
pop_df["since_1790"] = pop_df.year - 1790

#add column that is the population in millions
pop_df["millions"] = pop_df.us_pop / 1000000

#rounds to first decimal 
pop_df.millions = pop_df.millions.round(1)

#Define X and Y Values
x = pop_df.since_1790
y = pop_df.millions
```

### b)

```python
#Plot US population versus the years since 1790
plt.plot(x, y, label = 'Population in Millions')
plt.title('US Population versus Years since 1790')
plt.xlabel('Years since 1790')
plt.ylabel('Population in Millions')
plt.legend()
plt.show()
```

### c)

```python
#Create a Linear Regression Model to predict the US population in t years from 1790.

#Convert each dataframe into a numpy array
X = np.array(x).reshape(-1,1)
Y = np.array(y).reshape(-1,1)

#Create Linear regression model to predict population
reg = LinearRegression().fit(X,Y)
p1 = reg.predict(X)

#Print R-Squared Value
print(f'The R-Squared Value is: {reg.score(X,Y)}')
```

```python
#Adds a new column to square the number of years since 1790.
pop_df["sq_years"] = pop_df.since_1790 ** 2
pop_df.head()
```

### f)

```python
#Data plot
plt.scatter(x, y, label = 'Population in Millions')

#Plot 1st Model
plt.plot(x,p1,color='red', label ='Years since 1790')

#Plot 2nd Model
plt.plot(x,p2,color='green', label = 'Years squared')

plt.xlabel('Years since 1790')
plt.ylabel('Population in Millions')

plt.show()
```
