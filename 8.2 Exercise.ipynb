### 1. Finding the Eigenvalues and Eigenvectors of a Matrix

```python
#imports libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

a)

```python
A = np.array([[4,0,1],
            [-1,-6,-2],
            [5,0,0]])

u = np.array([[1],[2],[3]])

v = A @ u

plt.quiver(0, 0, u[0], u[1], color="#2EBCE7", angles='xy',
scale_units='xy', scale=1)
plt.quiver(0, 0, v[0], v[1], color="#00E64E", angles='xy',
scale_units='xy', scale=1)
print("This vector is NOT an eigenvector of A, because its not on the same line.")
```

b)

```python
x = np.array([[0],[1],[0]])
y = A @ x
plt.quiver(0, 0, x[0], x[1], color="#2EBCE7", angles='xy',
scale_units='xy', scale=1)
plt.quiver(0, 0, y[0], y[1], color="#00E64E", angles='xy',
scale_units='xy', scale=1)
print("This IS a eigenvecotr of A, because it is on the same line. ")
```

c)

```python
#find eigenvalues and eigenvectors
w,v = np.linalg.eig(A)

#printing eigenvalues
print("All the eigenvalues of A are", w)

#printing eigenvectors
print("All the eigenvectors of A are",v)
```
